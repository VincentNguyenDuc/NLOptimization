## Generating Dataset

We fit a nonlinear model $y = f(x;\beta) + \epsilon$ to a given dataset below, where $f(x;\beta) = \beta_0\cdot\exp(x/\beta_1)$.

```python
import numpy as np

def my_model(x, beta):
    return beta[0] * np.sin(beta[1] * x) + beta[2] * x**2

def generate_data(x, truth_beta, mean=0, std=0.1):
    # Generate y values with some noise
    y = my_model(x, truth_beta) + np.random.normal(mean, std, size=len(x))
    return y

n = 100
truth_beta = [1, 2, 3]
x = np.random.uniform(-10000, 10000, size=100)
y = generate_data(x, truth_beta, mean=0.5, std=1000)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, 'b*', label='data')
```
